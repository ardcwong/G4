# 6. Replace 'Male' and 'Female' in `gender` by 'M' and 'F'

# 7. Remove "$" in `amt` and convert to float
*Hint: convert to desired data type using method `.astype()`*

```python
df.info()
```

```python
df['amt'] = df['amt'].str.replace('$', '')
df['amt'] = df['amt'].astype('float')
```

# 8. Remove "people" and "," in `city_pop`and convert to integer
*Hint: convert to desired data type using method `.astype()`*

```python
df.info()
```

```python
df['city_pop'] = df['city_pop'].str.replace('people','')
df['city_pop'] = df['city_pop'].str.replace(',','')
```

```python
df.info()
```

```python
df['city_pop'] = df['city_pop'].astype('int')
```

```python
df.info()
```

# **Optional, For intermediate/advanced learners**:
## Column `unix_time` is a date unique system for machines to internally tracking time. Can you convert it to a human-readable date and time in "YYYY-MM-DD HH:MM:SS" format and store in column named 'transaction_datetime`?




```python
df['transaction_datetime'] = pd.to_datetime(df['unix_time'], unit = 's')
```

```python
df['transaction_datetime'] = df['transaction_datetime'].dt.tz_convert('Asia/Shanghai')
```

```python
df['transaction_datetime'] = df['transaction_datetime'].dt.strftime('%Y-%m-%d %H:%M:%S')
```
