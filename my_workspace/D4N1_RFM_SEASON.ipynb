## Prepare data

Read csv

```python
# Define the current date
current_date = pd.to_datetime('2022-01-01')
```

```python
df = pd.read_csv("Data/cc_clean.csv")
df.head()
```

```python
df.info()
```

```python
df['dob'] = pd.to_datetime(df['dob'], dayfirst = True)
```

```python
#AGE
year_diff = current_date.year - df['dob'].dt.year
had_birthday = (current_date.month > df['dob'].dt.month) | ((current_date.month == df['dob'].dt.month) & (current_date.day >= df['dob'].dt.day))
df['age'] = year_diff - (~had_birthday)
df.head()
```

*Trick*: Pandas has its own methods for dates. We just need to convert it using `pd.to_datetime`.

```python
#TIME OF DAY
df['trans_datetime'] = pd.to_datetime(df['trans_datetime'])
df['time_of_day'] = df['trans_datetime'].dt.hour
df['time_of_day'] = df['time_of_day'].map(lambda x: "morning" if (x >= 5 and x<12) else
                                                    "afternoon" if (x >= 12 and x < 17) else
                                                    "evening" if (x >= 17 and x < 21) else
                                                    "night")
df.head()

```

```python
df['season'] = df['trans_datetime'].dt.month
df['season'] = df['season'].map(lambda x: "Summer" if x in[3, 4, 5] else
                                                    "Holiday_Shopping" if x in [ 9, 10,11,12,1,2] else
                                                    "Rainy" )
df.head()
```

```python
def age_segment(age):
    if 50 <= age <= 59:
        return 'Middle-aged Adults'
    elif 60 <= age <= 69:
        return 'Pre-Retirees'
    elif 70 <= age <= 79:
        return 'Young Seniors'
    else:
        return 'Older Seniors'
```

```python
df['age_group'] = df['age'].map(age_segment)
df.head()
```

For recency analysis, we need to know how much days have passed since the transaction, assuming current date is Jan 1, 2022.

```python
# Calculate the elapsed days
df['elapsed_days'] = (current_date - df['trans_datetime']).dt.days
df.head()
```

## RFM Example
Determine some spending characteristics of each customer in the past 12 months


First, create a per-customer table using transactions done in 2021

```python
# Filter data to 2021
data_df = df[df['trans_datetime'].dt.year==2021]
data_df
```

### Recency


```python
# Get days since last transaction of each customer first, and include gender tag
# last transaction min of elapsed time
# groupby + trick to convert back to dataframe
rec_df = data_df.groupby(['acct_num','season', 'category'])['elapsed_days'].min().to_frame(name='recency').reset_index()
rec_df
```

Using this, create a system to assign a recency score to the `recency` column.

**More recent transaction => Higher score**

* 1 - low score
* 2 - medium score
* 3 - high score

```python
def assign_score_recency(x):
  if x<=30: # last transaction is within last month
    return 3
  elif 30<x<90: # within last quarter
    return 2
  else: #older than last quarter
    return 1
```

```python
rec_df['recency_score'] = rec_df['recency'].map(assign_score_recency)
rec_df
```

```python
rec_df.groupby(rec_df['recency_score'], rec_df['season'] == 'Holiday').nunique()
```

### Frequency
>Q: On average, how frequent did each customer  transact in the past year?

```python
# Get transaction frequency of each customer first
# groupby + trick to convert back to dataframe
freq_df = data_df.groupby(['acct_num','season', 'category']).size().to_frame(name='frequency').reset_index()
freq_df
```

Using pandas method `pd.qcut` (quantile cut), classify and score the customers based on their frequency

**More frequent transactions => Higher score**

* 1- low score
* 2- medium score
* 3- high score

```python
freq_df['frequency'].describe()
```

```python
freq_df['frequency_score'] = pd.qcut(freq_df['frequency'], q=3, labels=[1, 2, 3]).astype(int)
freq_df
```

```python
freq_df['frequency_score'].value_counts()
```

### Monetary value
>Q: On average, how much is the total spending of each customer per gender?

```python
# Get total spending of each customer first
# groupby + trick to convert back to dataframe
mv_df = data_df.groupby(['acct_num','season', 'category'])['amt'].sum().to_frame(name='total_amt').reset_index()
mv_df
```

Using pandas method `pd.qcut` (quantile cut), classify and score the customers based on their monetary value

**Larger total transaction amts => Higher score**

* 1- low score
* 2- medium score
* 3- high score

```python
mv_df['monetary_score'] = pd.qcut(mv_df['total_amt'], q=3, labels=[1, 2, 3]).astype(int)
mv_df
```

```python
mv_df['monetary_score'].value_counts()
```

```python
mv_df.groupby('monetary_score')['total_amt'].mean()
```

### Total RFM score
To compute the total score, we need to


1.   Merge the recency, frequency, and monetary value tables together
2.   Add the scores to a column `rfm_score`



```python
# First, merge rec_df and freq_df using pd.merge
rfm_df = pd.merge(rec_df, freq_df, on=['acct_num','season', 'category'], how='inner')
rfm_df
```

```python
print(len(rec_df))
print(len(freq_df))
print(len(rfm_df))
```

Add up the scores and store in column `rfm_score`

```python
rfm_df = pd.merge(rfm_df, mv_df, on=['acct_num','season', 'category'], how='inner')
rfm_df.head()
```

```python
rfm_df['rfm_score'] = rfm_df['recency_score'] + rfm_df['frequency_score'] +rfm_df['monetary_score']
```

```python
rfm_df['rfm_score'].value_counts()
```

```python
rfm_df
```

Create function to assign RFM levels

```python
def assign_rfm_level(x):
    if x >= 7:
        return 'Top'
    elif 4<=x<7:
        return 'Middle'
    else:
        return 'Low'
```

```python
rfm_df['rfm_level'] = rfm_df['rfm_score'].map(assign_rfm_level)
rfm_df.head()
```

```python
rfm_df
```

```python
rfm_summary = rfm_df.groupby(['season','category']).agg({'recency_score': 'mean', 'frequency_score': 'mean','monetary_score': 'mean'}).reset_index()
```

```python
rfm_summary['rfm_score'] = rfm_summary['recency_score'] + rfm_summary['frequency_score'] +rfm_summary['monetary_score']
rfm_summary['rfm_level'] = rfm_summary['rfm_score'].map(assign_rfm_level)
rfm_summary.sort_values("rfm_score", ascending = False)
```

```python
# Helper function to plot RFM metrics
def plot_rfm_metrics(rfm_summary, metric, title):
   plt.figure(figsize=(14, 8))
   sns.barplot(data=rfm_summary, x='season', y=(metric), hue='category')
   plt.title(title)
   plt.ylabel(f'Mean {metric.capitalize()}')
   plt.xlabel('Season')
   plt.legend(title='Category', bbox_to_anchor=(1.05, 1), loc='upper left')
   plt.show()
# Plot mean recency, frequency, and monetary for each time_of_day_segment and category
plot_rfm_metrics(rfm_summary, 'recency_score', 'Mean Recency by Season and Category')
plot_rfm_metrics(rfm_summary, 'frequency_score', 'Mean Frequency by Season and Category')
plot_rfm_metrics(rfm_summary, 'monetary_score', 'Mean Monetary Value by Season and Category')
```
